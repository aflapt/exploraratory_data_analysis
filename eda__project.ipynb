## Find out under which team and position, spending in terms of salary is high.

```python
x = df_employee["Salary"].mean()
df_employee["Salary"].fillna(x,inplace =True)
df_employee
```

```python
df_employee["Salary"].value_counts()
```

```python
df1 = df_employee[["Name","Team","Position","Salary"]]
df1
```

```python
#find maxmum salary
total_salary = df1.groupby(["Team","Position"])["Salary"].sum()
max_spending = total_salary.idxmax()
max_spending_salary = total_salary.max()
print(f"The team and position with the highest spending is:\n Team: {max_spending[0]}\n position: {max_spending[1]} \n with a total salary of {max_spending_salary}.")
```

```python
plt.figure(figsize=(20,5))
sns.scatterplot(y = "Salary",x="Team",hue = "Position",data = df_employee)
plt.ylabel("Salary")
plt.xlabel("Team")
plt.title("scatter plot position vs team with salary")
plt.xticks(rotation=45)
plt.show()
```

**data story**:
scatter plot gives salary distribution accross different position within each team.it reveal that Los angeles lakers satndout highest spending team






## Find if there is any correlation between age and salary , represent it visually

**correlation**:

Values close to +1 indicates strong positive correlation, -1 indicates a strong negative correlation and 0 indicates suggests no linear correlation.
Darker colors signify strong correlation, while light colors represents weaker correlations.
Positive correlation variable move in same directions. As one increases, the other also increases.
Negative correlation variable move in opposite directions. An increase in one variable is associated with a decrease in the other.


```python
df2 = df_employee[["Age","Salary"]]
df2
```

```python
corr = df2.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="viridis")
plt.title("correlation matrix between age and salary of employees")
plt.show()
```

**data story**:
The correlation shows the relationship between "Age" and "Salary" in the employee dataset:

**age vs age(1.0000)**0) : This value of 1.00000 along the diagonal represents the correlation of a variable with itself, which is always 1. It indicates a perfect positive correlation, as expecte
d.**

Age vs. Salary (0.21**117) : The correlation coefficient of 0.21117 suggests a positive but relatively weak relationship between age and salary. This means that as age increases, there's a tendency for salaries to increase slightly, but the correlation isn't very str
ong.

Simil**arly, Salary vs. Age (0.**21117) : Since correlation is a symmetric measure, this value mirrors the correlation between "Age" and "Salary" and reaffirms the weak positive relationship between the two vari
ables.

In simpler terms, the correlation matrix indicates that there's a positive but not particularly strong relationship between age and salary in the employee dataset. As employees' ages tend to increase, there is a tendency for their salaries to slightly rise.
